```python
from pdf2image import convert_from_path
import pytesseract
import os
```

```python
def ocr_pdf(pdf_path, lang_code="eng"):
    #convert PDF to image list
    images = convert_from_path(pdf_path)
    full_text = ""
    for i, image in enumerate(images):
        text = pytesseract.image_to_string(image, lang=lang_code)
        full_text += text + "\n"
    return full_text.strip()
```

```python
import re
from nltk.tokenize import sent_tokenize
import nltk
nltk.download('punkt')
```

```python
def fixed_length_chunking(text, chunk_size=500):
    return [text[i:i+chunk_size] for i in range(0, len(text), chunk_size)]
```

```python
def overlapping_sentence_chunks(text, max_sentences=5, overlap=2):
    sentences = sent_tokenize(text)
    chunks = []
    i = 0
    while i < len(sentences):
        chunk = ' '.join(sentences[i:i+max_sentences])
        chunks.append(chunk)
        i += max_sentences - overlap
    return chunks
```

```python
# For English
english_fixed_chunks = fixed_length_chunking(english_text, chunk_size=500)

# Download the specific punkt_tab resource required by sent_tokenize
nltk.download('punkt_tab')

english_sentence_chunks = overlapping_sentence_chunks(english_text, max_sentences=5, overlap=2)

# For Urdu (same strategy, assuming Urdu OCR works well with sentence tokenization)
# You might also need to download the punkt_tab resource for 'urdu' if sent_tokenize is used with lang='urdu'
# nltk.download('punkt_tab', lang='urdu') # Uncomment if needed for Urdu

urdu_fixed_chunks = fixed_length_chunking(urdu_text, chunk_size=500)
urdu_sentence_chunks = overlapping_sentence_chunks(urdu_text, max_sentences=5, overlap=2)

print("Sample English Fixed Chunk:\n", english_fixed_chunks[0])
print("\nSample Urdu Overlapping Chunk:\n", urdu_sentence_chunks[0])
```

```python
models = {
    "e5": SentenceTransformer("intfloat/multilingual-e5-large"),
    "gte": SentenceTransformer("thenlper/gte-large"),
    "miniLM": SentenceTransformer("sentence-transformers/paraphrase-multilingual-MiniLM-L12-v2")
}
```

```python
def fixed_length_chunking(text, chunk_size=300):
    return [text[i:i+chunk_size] for i in range(0, len(text), chunk_size)]
```

```python
import nltk
from nltk.tokenize import sent_tokenize
```

```python
nltk.download('punkt')
```

```python
def sentence_based_chunking(text, max_sentences=3):
    sentences = sent_tokenize(text)
    chunks = []
    for i in range(0, len(sentences), max_sentences):
        chunk = " ".join(sentences[i:i+max_sentences])
        chunks.append(chunk)
    return chunks
```

```python
chunks_fixed_en = fixed_length_chunking(english_text, chunk_size=300)
chunks_sent_en = sentence_based_chunking(english_text, max_sentences=3)

chunks_fixed_ur = fixed_length_chunking(urdu_text, chunk_size=300)
chunks_sent_ur = sentence_based_chunking(urdu_text, max_sentences=3)

# Combine all into a single list
chunks = chunks_fixed_en + chunks_sent_en + chunks_fixed_ur + chunks_sent_ur

print(f"✅ Total chunks prepared: {len(chunks)}")
```

```python
embedding_results = {}
```

```python
for model_name, model in models.items():
    print(f"\nGenerating embeddings using: {model_name}")

    embeddings = model.encode(
        chunks,
        convert_to_tensor=True,
        show_progress_bar=True,
        normalize_embeddings=True  # normalize for cosine similarity
    )

    embedding_results[model_name] = {
        "chunks": chunks,
        "embeddings": embeddings
    }
```
